```python
import pandas as pd
import plotly.express as px


df = pd.read_csv('https://raw.githubusercontent.com/pattaradanai-lks/Plotly/main/forbes2000-2022.csv')
df.head()
```

```python
df.shape
```

```python
# ตัดเครื่องหมาย $ และ B,M ออกจากคอลัมน์ที่ต้องการ เพื่อต้องการเรียงลำดับข้อมูลเป็นตัวเลข
df['sales '] = df['sales '].astype(str).str.replace('$', '').str.replace('B', '').str.replace('M', '').astype(float)
df['profit '] = df['profit '].astype(str).str.replace('$', '').str.replace('B', '').str.replace('M', '').astype(float)
df['market value'] = df['market value'].astype(str).str.replace('$', '').str.replace('B', '').str.replace(',', '').astype(float)
```

```python
# เรียงลำดับข้อมูลตามคอลัมน์ 'market value' จากมากไปน้อย
df = df.sort_values(by='market value', ascending=False)

```

```python
df.head(50)
```

```python
df=df.head(50)
```

```python
df.columns
```

```python
fig = px.scatter(data_frame=df, x='profit ', y='sales ',
                 size='market value', color="country",
                 hover_name="global company", size_max=80,
                 labels={'profit ': 'Profit (Billion)$', 'sales ': 'Sales (Billion)$','market value':'Market value (Billion)$'})
fig.show()
```
